```python
import pandas as pd
```

```python
df = pd.read_csv('../input/allholds.csv', sep='|', compression='gzip')
```

```python
df['charge_release_dt'] = pd.to_datetime(df['charge_release_dt'])
```

```python
len(set(df['booking_id']))
```

```python
no_charge = df['charge_count'] == 0
```

```python
len(set(df.loc[no_charge,'booking_id']))
```

```python
df.loc[no_charge]
```

```python
last_released_idx = df.groupby('booking_id')['charge_release_dt'].idxmax()

last_released_charge = df.loc[last_released_idx]['release_disposition_desc']

last_released_idx = last_released_idx.reset_index().set_index('charge_release_dt')

temp = last_released_idx.join(last_released_charge)

temp.reset_index(inplace=True)

temp['final_release_disp'] = temp['release_disposition_desc']

temp.drop(['charge_release_dt', 'release_disposition_desc'], axis=1, inplace=True)

df = pd.merge(df, temp, on='booking_id', how='left')
```

```python
df.drop_duplicates(subset='booking_id')['final_release_disp'].value_counts()
```
