```python
import pandas as pd
```

```python
df = pd.read_csv('../input/pierce.csv.gz', sep='|', compression='gzip')
```

```python
bail_2013_2017 = pd.read_csv('../input/bookings_with_bail.csv', sep='|')
bail_2017_2018 = pd.read_csv('../input/Bail_Receipt_List_Rpt.csv', sep='|')
```

```python
bail_2013_2017.columns == bail_2017_2018.columns
```

```python
bail_2013_2017.info()
```

```python
bail_2017_2018
```

```python
bail = pd.concat([bail_2013_2017, bail_2017_2018])
```

```python
bail.tail()
```

```python
types = [type(x) for x in bail['fine_amt']]
```

```python
types[0] == float
```

```python
bail['booking_charge_cd_desc'] = bail['booking_charge_desc'].copy()
```

```python
bail['booking_charge_desc'] = bail['booking_charge_cd_desc'].str.split(' - ', 1).str[1]
```

```python
bailcols = ['booking_id',
            'bail_receipt_id',
            'bail_receipt_dt',
            'booking_charge_desc',
            'bail_fund_type_cd',
            'bail_fund_type_desc',
            'fine_amt',
            'start_receipt_id',
            'end_receipt_id',
            'disbursed_status',
            'disbursed_dt']
```

```python
merge_cols = ['booking_id', 'booking_charge_desc']
```

```python
df = pd.merge(df, bail[bailcols], on=merge_cols, how='left')
```

```python
bookings_w_bail = set(df.dropna(subset=['bail_receipt_id'])['booking_id'])
```

```python
bail_df = df[df['booking_id'].isin(bookings_w_bail)]
```

```python
bail_df.info()
```

```python
bail_df.to_csv('../frozen/bail.csv')
```
