```python
import pandas as pd
```

```python
charges = pd.read_csv('../hand/charges20181005.csv')
```

```python
pierce_charges = charges[charges['jurisdiction'] == 'pierce']
```

```python
pierce_charges
```

```python
df = pd.read_csv('../input/pierce.csv.gz', sep='|', compression='gzip')
```

```python
value_counts = df['booking_charge_desc'].value_counts()
```

```python
value_counts.name = 'charge_count'
```

```python
pierce_charges = pierce_charges.join(value_counts, on='booking_charge_desc')
```

```python
pierce_charges.to_csv('../hand/charges20181005.csv')
```
